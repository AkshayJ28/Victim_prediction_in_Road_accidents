```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('cleaned_data.csv')
df.shape
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder 
  
le = LabelEncoder() 
  
df['Accident cause']= le.fit_transform(df['Accident cause'].astype(str)) 
df['Accident classfication']= le.fit_transform(df['Accident classfication'].astype(str)) 
df['Climate']= le.fit_transform(df['Climate'].astype(str)) 
df['Weekday']= le.fit_transform(df['Weekday']) 
df['Moment of the day']= le.fit_transform(df['Moment of the day'].astype(str)) 
df['city']= le.fit_transform(df['city'].astype(str)) 
df['regional']= le.fit_transform(df['regional'].astype(str)) 
df['Road way']= le.fit_transform(df['Road way'].astype(str)) 
df['Accident type']= le.fit_transform(df['Accident type'].astype(str)) 
df['Road type']= le.fit_transform(df['Road type'].astype(str))
df['Road layout']= le.fit_transform(df['Road layout'].astype(str)) 
df['State']= le.fit_transform(df['State'].astype(str))
df['Soil use']= le.fit_transform(df['Soil use'].astype(str))
```
